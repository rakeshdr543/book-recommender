```python
books["tagged_description"].to_csv("tagged_description.txt",
                                   sep = "\n",
                                   index = False,
                                   header = False)
```

```python
from langchain.schema import Document

with open("tagged_description.txt", "r", encoding="utf-8") as f:
    lines = f.readlines()

documents = [Document(page_content=line.strip()) for line in lines if line.strip()]

print(f"Created {len(documents)} documents for Chroma")
```

```python
documents[0]
```

```python
db_books = Chroma.from_documents(
    documents,
    embedding=OpenAIEmbeddings())
```

```python
query = "A book to teach children about nature"
docs = db_books.similarity_search(query, k = 10)
docs
```

```python
query = "Investing and financial learn"
docs = db_books.similarity_search(query, k=10)
docs
```

```python
def retrieve_semantic_recommendations(
        query: str,
        top_k: int = 10,
) -> pd.DataFrame:
    recs = db_books.similarity_search(query, k = 50)

    books_list = []

    for i in range(0, len(recs)):
        books_list += [int(recs[i].page_content.strip('"').split()[0])]

    return books[books["isbn13"].isin(books_list)]
```

```python
retrieve_semantic_recommendations("A book to teach children about nature")
```
